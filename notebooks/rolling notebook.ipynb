* `match_id`

This table provides the **temporal structure** required to analyze player workload over time.

## 3. Events Data — Limitations

## Step 3 — Create Rolling Workload Features

To capture player workload and recovery dynamics, we compute time-based rolling features using past match data.

These features are calculated per player and only include information available **before the current match** (using `.shift(1)` to avoid data leakage).

### Workload Features

* **minutes_7d** → total minutes played in the previous 7 days (short-term / acute load)
* **minutes_14d** → total minutes played in the previous 14 days (medium-term load)
* **minutes_28d** → total minutes played in the previous 28 days (long-term / chronic load)

These features quantify how much physical load a player has accumulated over different time horizons.

---

### Match Congestion Features

* **matches_7d** → number of matches played in the previous 7 days
* **matches_14d** → number of matches played in the previous 14 days
* **matches_28d** → number of matches played in the previous 28 days

These features capture schedule congestion and frequency of match exposure.

---

### Recovery Feature

* **days_since_last_match** → number of days since the player’s previous match

This feature reflects recovery time and helps identify short rest periods, which are associated with increased injury risk.

---

### Key Principle — Time-Based Rolling Windows

Matches are not evenly spaced in time, so workload is computed using **time-based rolling windows (e.g., 7D, 14D, 28D)** rather than a fixed number of matches.

This approach better captures real-world conditions such as fixture congestion and irregular match schedules.

---

### Key Principle — No Data Leakage

All features are computed using only past data.

The `.shift(1)` operation ensures that the current match is excluded from the calculations, making the features suitable for predictive modeling.

#Convert Time to Minutes
Converts time from format "MM:SS" → numeric minutes

If to = None → assumes player stayed until end of match

Player minutes are computed using the 'from' and 'to' fields provided in the lineups data.

Each interval represents a period during which the player was on the pitch. Total playing time is obtained by summing all intervals for a given player in a match.

```python
def time_to_minutes(t):
    if t is None:
        return 95  # approximate end of match (includes extra time)
    minutes, seconds = map(int, t.split(':'))
    return minutes + seconds / 60
```

# Extract Player Minutes from Lineups
Loops through all matches

Opens lineup file for each match

For each player:

Reads all time intervals (from → to)

Converts them into minutes

Sums them to get total playing time

Output:1 row = 1 player in 1 match with minutes played

```python
players_data = []

for match_id in matches_df['match_id']:
    file_path = f"{lineups_path}/{match_id}.json"

    if not os.path.exists(file_path):
        continue

    with open(file_path) as f:
        lineup = json.load(f)

    for team in lineup:
        for player in team['lineup']:

            total_minutes = 0

            for pos in player['positions']:
                start = time_to_minutes(pos['from'])
                end = time_to_minutes(pos['to'])

                total_minutes += (end - start)

            players_data.append({
                'match_id': match_id,
                'player_id': player['player_id'],
                'player_name': player['player_name'],
                'minutes_played': total_minutes
            })

lineup_minutes_df = pd.DataFrame(players_data)
```

# Merge with Match Dates
Adds the match date to each player entry

Converts dates into proper datetime format
required for for time-based features (rolling windows).

---



```python
final_df = lineup_minutes_df.merge(
    matches_df[['match_id', 'match_date']],
    on='match_id',
    how='left'
)

final_df['match_date'] = pd.to_datetime(final_df['match_date'])
```

#Rolling feartures


# Sort Data Chronologically
The dataset is first sorted by player_id and match_date.

This ensures that each player’s matches are ordered in time, which is necessary to correctly compute past workload and recovery features.

```python
final_df = final_df.sort_values(['player_id', 'match_date'])
```

# Days Since Last Match
For each player, we compute the number of days between consecutive matches.

This is done using the difference between match dates.

This feature captures recovery time and helps identify short rest periods, which are associated with higher injury risk.

```python
final_df['days_since_last_match'] = (
    final_df.groupby('player_id')['match_date']
    .diff()
    .dt.days
)
```

# Rolling Minutes (7 Days)

For each player and each match, we compute the total number of minutes played in the previous 7 days.

This is done using a time-based rolling window (7D).

Importantly, the current match is excluded using .shift(1) to ensure that only past information is used.

This feature represents short-term workload.

```python
final_df = final_df.sort_values(['player_id', 'match_date']).reset_index(drop=True)

def compute_minutes_7d(group):
    group = group.set_index('match_date')
    group['minutes_7d'] = group['minutes_played'].rolling('7D').sum().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_minutes_7d)

final_df = final_df.reset_index()
```

# Rolling Minutes (14 Days)

Similarly, we compute the total minutes played in the previous 14 days.

This captures medium-term workload and provides a broader view of player load accumulation.

```python
def compute_minutes_14d(group):
    group = group.set_index('match_date')
    group['minutes_14d'] = group['minutes_played'].rolling('14D').sum().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_minutes_14d)
final_df = final_df.reset_index()
```

# Rolling Minutes (28 Days)
We compute the total number of minutes played in the previous 28 days.

This represents long-term workload (chronic load) and is commonly used as a baseline to assess whether a player is experiencing unusual increases in activity.

As with other rolling features, the current match is excluded using `.shift(1)` to avoid data leakage.

```python
def compute_minutes_28d(group):
    group = group.set_index('match_date')
    group['minutes_28d'] = group['minutes_played'].rolling('28D').sum().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_minutes_28d)
final_df = final_df.reset_index()
```

# Number of Matches (7 Days)

We count how many matches each player has played in the previous 7 days.

This is a measure of schedule congestion and reflects how frequently a player is exposed to match intensity.

```python
def compute_matches_7d(group):
    group = group.set_index('match_date')
    group['matches_7d'] = group['minutes_played'].rolling('7D').count().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_matches_7d)
final_df = final_df.reset_index()
```

## Matches Played (14 Days)

We compute the number of matches played in the previous 14 days.

This feature captures medium-term schedule congestion and reflects how frequently a player has been exposed to match intensity.

As with other rolling features, the current match is excluded using `.shift(1)` to avoid data leakage.

```python
def compute_matches_14d(group):
    group = group.set_index('match_date')
    group['matches_14d'] = group['minutes_played'].rolling('14D').count().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_matches_14d)
final_df = final_df.reset_index()
```

## Matches Played (28 Days)

We compute the number of matches played in the previous 28 days.

This represents long-term match exposure and helps identify sustained periods of high competition load.

The current match is excluded using `.shift(1)` to ensure only past data is used.

---



```python
def compute_matches_28d(group):
    group = group.set_index('match_date')
    group['matches_28d'] = group['minutes_played'].rolling('28D').count().shift(1)
    return group

final_df = final_df.groupby('player_id', group_keys=False).apply(compute_matches_28d)
final_df = final_df.reset_index()
```

# Handling Missing Values

For the first matches of each player, there is no prior history, which results in missing values.

These are replaced with 0, indicating no previous workload.

```python
final_df[
    ['minutes_7d', 'minutes_14d', 'minutes_28d',
     'matches_7d', 'matches_14d', 'matches_28d']
] = final_df[
    ['minutes_7d', 'minutes_14d', 'minutes_28d',
     'matches_7d', 'matches_14d', 'matches_28d']
].fillna(0)
```
